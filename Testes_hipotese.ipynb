```python
# Instancia 01
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv(r'C:\FIAP\Hipoteses_Nemec_09_03_26_Futebol\Dados\instancia_01_futebol\dataset.csv')

cor = df[df['time'] == 'Corinthians']
pal = df[df['time'] == 'Palmeiras']

# A)
t, p = stats.ttest_ind(cor['finalizacoes'], pal['finalizacoes'], equal_var=False)
media_cor = cor['finalizacoes'].mean()
media_pal = pal['finalizacoes'].mean()
media_geral = df['finalizacoes'].mean()
print(f"Média Corinthians: {media_cor:.2f}")
print(f"Média Palmeiras:   {media_pal:.2f}")
print(f"t = {t:.2f}, p = {p:.2f}")

if p < 0.05:
    vencedor = 'Palmeiras' if media_pal > media_cor else 'Corinthians'
    resp_a = max(media_cor, media_pal)
    print(f"Diferença SIGNIFICATIVA → {vencedor} finaliza mais")
    print(f"Resposta A = {resp_a:.2f}")
else:
    resp_a = media_geral
    print(f"Sem diferença significativa → média geral = {media_geral:.4f}")
    print(f"Resposta A = {resp_a:.2f}")

# B)
t2, p2 = stats.ttest_ind(cor['faltas'], pal['faltas'], equal_var=True)
media_cor_f = cor['faltas'].mean()
media_pal_f = pal['faltas'].mean()
media_geral_f = df['faltas'].mean()
print(f"Média Corinthians: {media_cor_f:.2f}")
print(f"Média Palmeiras:   {media_pal_f:.2f}")
print(f"t = {t2:.2f}, p = {p2:.2f}")

if p2 < 0.05:
    vencedor_f = 'Palmeiras' if media_pal_f > media_cor_f else 'Corinthians'
    resp_b = max(media_cor_f, media_pal_f)
    print(f"Diferença SIGNIFICATIVA → {vencedor_f} faz mais faltas")
    print(f"Resposta B = {resp_b:.2f}")
else:
    resp_b = media_geral_f
    print(f"Sem diferença significativa → média geral = {media_geral_f:.4f}")
    print(f"Resposta B = {resp_b:.2f}")

# C)
u, p3 = stats.mannwhitneyu(cor['cartoes_amarelos'], pal['cartoes_amarelos'], alternative='two-sided')
med_cor_c = cor['cartoes_amarelos'].median()
med_pal_c = pal['cartoes_amarelos'].median()
med_geral_c = df['cartoes_amarelos'].median()
print(f"Mediana Corinthians: {med_cor_c:.2f}")
print(f"Mediana Palmeiras:   {med_pal_c:.2f}")
print(f"U = {u:.2f}, p = {p3:.2f}")

if p3 < 0.05:
    vencedor_c = 'Palmeiras' if med_pal_c > med_cor_c else 'Corinthians'
    resp_c = max(med_cor_c, med_pal_c)
    print(f"Diferença SIGNIFICATIVA → {vencedor_c} recebe mais cartões")
    print(f"Resposta C = {resp_c:.2f}")
else:
    resp_c = med_geral_c
    print(f"Sem diferença significativa → mediana geral = {med_geral_c:.2f}")
    print(f"Resposta C = {resp_c:.2f}")

# D)
cor_sorted = cor.sort_values('rodada').reset_index(drop=True)
n = len(cor_sorted)
metade = n // 2
primeira = cor_sorted.iloc[:metade]['gols'].values
segunda  = cor_sorted.iloc[metade:]['gols'].values
# Garantir mesmo tamanho para o teste pareado
min_len = min(len(primeira), len(segunda))
primeira = primeira[:min_len]
segunda  = segunda[:min_len]
t4, p4 = stats.ttest_rel(primeira, segunda)
media_1 = primeira.mean()
media_2 = segunda.mean()
media_geral_cor = cor['gols'].mean()
print(f"n total Corinthians: {n}, metade: {metade}")
print(f"Média 1ª metade: {media_1:.4f}")
print(f"Média 2ª metade: {media_2:.4f}")
print(f"t = {t4:.4f}, p = {p4:.4f}")

if p4 < 0.05:
    melhor = '1ª metade' if media_1 > media_2 else '2ª metade'
    resp_d = max(media_1, media_2)
    print(f"Diferença SIGNIFICATIVA → {melhor} tem mais gols")
    print(f"Resposta D = {resp_d:.2f}")
else:
    resp_d = media_geral_cor
    print(f"Sem diferença significativa → média geral Corinthians = {media_geral_cor:.4f}")
    print(f"Resposta D = {resp_d:.2f}")

print()
print("="*60)
print("RESUMO FINAL")
print("="*60)
print(f"A = {resp_a:.2f}")
print(f"B = {resp_b:.2f}")
print(f"C = {resp_c:.2f}")
print(f"D = {resp_d:.2f}")
total = resp_a + resp_b + resp_c + resp_d
print(f"\nSOMA TOTAL = {total:.2f}")
```

```python
# Instancia 2

df = pd.read_csv(r'C:\FIAP\Hipoteses_Nemec_09_03_26_Futebol\Dados\instancia_02_futebol\dataset.csv')
print(df.shape)
df.head()

# A)

cor = df[df['time'] == 'Corinthians']['cartoes_amarelos']
pal = df[df['time'] == 'Palmeiras']['cartoes_amarelos']

stat_a, p_a = stats.mannwhitneyu(cor, pal, alternative='two-sided')

med_cor = cor.median()
med_pal = pal.median()
med_geral = df['cartoes_amarelos'].median()

print(f"Mann-Whitney U = {stat_a:.4f} | p-value = {p_a:.4f}")
print(f"Mediana Corinthians: {med_cor} | Mediana Palmeiras: {med_pal}")
print(f"Mediana geral: {med_geral}")

if p_a < 0.05:
    resultado_a = max(med_cor, med_pal)
    vencedor = 'Corinthians' if med_cor > med_pal else 'Palmeiras'
    print(f"\nDiferença SIGNIFICATIVA → Maior mediana: {vencedor} = {resultado_a}")
else:
    resultado_a = med_geral
    print(f"\n Sem diferença significativa → Mediana geral = {resultado_a}")

resultado_a = round(resultado_a, 2)
print(f"\nRESULTADO ITEM A = {resultado_a}")

# B)

cor_all = df[df['time'] == 'Corinthians'].sort_values('rodada').reset_index(drop=True)

n = len(cor_all)
metade = n // 2

primeira_metade = cor_all.iloc[:metade]['gols'].values
segunda_metade  = cor_all.iloc[metade:]['gols'].values

stat_b, p_b = stats.ttest_rel(primeira_metade, segunda_metade)

media_1 = primeira_metade.mean()
media_2 = segunda_metade.mean()
media_geral_cor = cor_all['gols'].mean()

print(f"t-stat = {stat_b:.4f} | p-value = {p_b:.4f}")
print(f"Média 1ª metade: {media_1:.4f} | Média 2ª metade: {media_2:.4f}")
print(f"Média geral Corinthians: {media_geral_cor:.4f}")

if p_b < 0.05:
    resultado_b = max(media_1, media_2)
    melhor = '1ª metade' if media_1 > media_2 else '2ª metade'
    print(f"\nDiferença SIGNIFICATIVA → Melhor média: {melhor} = {resultado_b:.4f}")
else:
    resultado_b = media_geral_cor
    print(f"\nSem diferença significativa → Média geral Corinthians = {resultado_b:.4f}")

resultado_b = round(resultado_b, 2)
print(f"\nRESULTADO ITEM B = {resultado_b}")

# C)

cor_all = df[df['time'] == 'Corinthians'].sort_values('rodada').reset_index(drop=True)

n = len(cor_all)
metade = n // 2

posse_1 = cor_all.iloc[:metade]['posse_bola'].values
posse_2 = cor_all.iloc[metade:]['posse_bola'].values

stat_c, p_c = stats.wilcoxon(posse_1, posse_2, alternative='two-sided')

med_1 = np.median(posse_1)
med_2 = np.median(posse_2)
med_geral_cor = cor_all['posse_bola'].median()

print(f"Wilcoxon stat = {stat_c:.4f} | p-value = {p_c:.4f}")
print(f"Mediana 1ª metade: {med_1:.4f} | Mediana 2ª metade: {med_2:.4f}")
print(f"Mediana geral Corinthians: {med_geral_cor:.4f}")

if p_c < 0.05:
    resultado_c = max(med_1, med_2)
    maior = '1ª metade' if med_1 > med_2 else '2ª metade'
    print(f"\nDiferença SIGNIFICATIVA → Maior mediana: {maior} = {resultado_c:.4f}")
else:
    resultado_c = med_geral_cor
    print(f"\nSem diferença significativa → Mediana geral Corinthians = {resultado_c:.4f}")

resultado_c = round(resultado_c, 2)
print(f"\nRESULTADO ITEM C = {resultado_c}")

# D)

faltas_cor = df[df['time'] == 'Corinthians']['faltas']
faltas_pal = df[df['time'] == 'Palmeiras']['faltas']

# Teste de Levene para verificar homogeneidade de variâncias
stat_lev, p_lev = stats.levene(faltas_cor, faltas_pal)
equal_var = p_lev > 0.05
print(f"Levene: stat={stat_lev:.4f}, p={p_lev:.4f} → equal_var={equal_var}")

stat_d, p_d = stats.ttest_ind(faltas_cor, faltas_pal, equal_var=equal_var, alternative='two-sided')

media_cor = faltas_cor.mean()
media_pal = faltas_pal.mean()
media_geral = df[df['time'].isin(['Corinthians','Palmeiras'])]['faltas'].mean()

print(f"\nt-stat = {stat_d:.4f} | p-value = {p_d:.4f}")
print(f"Média faltas Corinthians: {media_cor:.4f} | Média faltas Palmeiras: {media_pal:.4f}")
print(f"Média geral (ambos): {media_geral:.4f}")

if p_d < 0.05:
    resultado_d = max(media_cor, media_pal)
    mais_faltoso = 'Corinthians' if media_cor > media_pal else 'Palmeiras'
    print(f"\nDiferença SIGNIFICATIVA → Mais faltoso: {mais_faltoso} com média = {resultado_d:.4f}")
else:
    resultado_d = media_geral
    print(f"\nSem diferença significativa → Média geral = {resultado_d:.4f}")

resultado_d = round(resultado_d, 2)
print(f"\nRESULTADO ITEM D = {resultado_d}")

soma = resultado_a + resultado_b + resultado_c + resultado_d

print("=" * 40)
print(f"  Resultado A: {resultado_a}")
print(f"  Resultado B: {resultado_b}")
print(f"  Resultado C: {resultado_c}")
print(f"  Resultado D: {resultado_d}")
print("=" * 40)
print(f"  SOMA FINAL = {round(soma, 2)}")
print("=" * 40)
```

```python
# Instancia 4

df = pd.read_csv(r'C:\FIAP\Hipoteses_Nemec_09_03_26_Futebol\Dados\instancia_04_ecommerce\dataset.csv')
alpha = 0.05

# A)
antes = df['conversao_antes_pct']
depois = df['conversao_depois_pct']

stat_a, p_a = stats.wilcoxon(antes, depois, alternative='two-sided')

med_antes  = np.median(antes)
med_depois = np.median(depois)
med_geral_a = np.median(pd.concat([antes, depois]))

print(f"Estatística W : {stat_a}")
print(f"p-value       : {p_a:.6f}")
print(f"Mediana antes : {med_antes:.2f}")
print(f"Mediana depois: {med_depois:.2f}")
print(f"Mediana geral : {med_geral_a:.2f}")

if p_a < alpha:
    resp_a = round(max(med_antes, med_depois), 2)
    print(f"→ Diferença SIGNIFICATIVA → maior mediana: {resp_a}")
else:
    resp_a = round(med_geral_a, 2)
    print(f"→ Sem diferença → mediana geral: {resp_a}")

# B)
sat_a = df[df['campanha'] == 'A']['satisfacao']
sat_b = df[df['campanha'] == 'B']['satisfacao']

stat_b, p_b = stats.mannwhitneyu(sat_a, sat_b, alternative='two-sided')

med_sa = np.median(sat_a)
med_sb = np.median(sat_b)
med_sg = np.median(df['satisfacao'])

print(f"Estatística U : {stat_b}")
print(f"p-value       : {p_b:.6f}")
print(f"Mediana A     : {med_sa:.2f}")
print(f"Mediana B     : {med_sb:.2f}")
print(f"Mediana geral : {med_sg:.2f}")

if p_b < alpha:
    resp_b = round(max(med_sa, med_sb), 2)
    print(f"→ Diferença SIGNIFICATIVA → maior mediana: {resp_b}")
else:
    resp_b = round(med_sg, 2)
    print(f"→ Sem diferença → mediana geral: {resp_b}")

# C)
mobile  = df[df['dispositivo'] == 'mobile']['tempo_antes_min']
desktop = df[df['dispositivo'] == 'desktop']['tempo_antes_min']

stat_c, p_c = stats.ttest_ind(mobile, desktop, equal_var=False, alternative='two-sided')

mean_mob  = np.mean(mobile)
mean_desk = np.mean(desktop)
mean_tg_c = np.mean(df['tempo_antes_min'])

print(f"Estatística t : {stat_c:.4f}")
print(f"p-value       : {p_c:.6f}")
print(f"Média mobile  : {mean_mob:.2f}")
print(f"Média desktop : {mean_desk:.2f}")
print(f"Média geral   : {mean_tg_c:.2f}")

if p_c < alpha:
    resp_c = round(max(mean_mob, mean_desk), 2)
    print(f"→ Diferença SIGNIFICATIVA → maior média: {resp_c}")
else:
    resp_c = round(mean_tg_c, 2)
    print(f"→ Sem diferença → média geral: {resp_c}")

# D)
tk_a = df[df['campanha'] == 'A']['ticket_medio']
tk_b = df[df['campanha'] == 'B']['ticket_medio']

stat_d, p_d = stats.ttest_ind(tk_a, tk_b, equal_var=True, alternative='two-sided')

mean_ta = np.mean(tk_a)
mean_tb = np.mean(tk_b)
mean_tg_d = np.mean(df['ticket_medio'])

print(f"Estatística t : {stat_d:.4f}")
print(f"p-value       : {p_d:.6f}")
print(f"Média A       : {mean_ta:.2f}")
print(f"Média B       : {mean_tb:.2f}")
print(f"Média geral   : {mean_tg_d:.2f}")

if p_d < alpha:
    resp_d = round(max(mean_ta, mean_tb), 2)
    print(f"→ Diferença SIGNIFICATIVA → maior ticket médio: {resp_d}")
else:
    resp_d = round(mean_tg_d, 2)
    print(f"→ Sem diferença → ticket médio geral: {resp_d}")

print("RESUMO DOS RESULTADOS")
print("=" * 60)
print(f"a) {resp_a}")
print(f"b) {resp_b}")
print(f"c) {resp_c}")
print(f"d) {resp_d}")
print(f"{'─'*30}")
print(f"SOMA FINAL: {round(resp_a + resp_b + resp_c + resp_d, 2)}")
```

```python
# Instancia 20
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv(r'C:\FIAP\Hipoteses_Nemec_09_03_26_Futebol\Dados\instancia_20_manufatura\dataset.csv')
df.columns = df.columns.str.strip().str.lstrip('\ufeff')

alpha = 0.05

m1 = df[df['maquina'] == 'M1']
m2 = df[df['maquina'] == 'M2']

# ITEM A - Vibração: Mann-Whitney U
vib_m1 = m1['vibracao']
vib_m2 = m2['vibracao']
u_stat, p_val_a = stats.mannwhitneyu(vib_m1, vib_m2, alternative='two-sided')
med_vib_m1 = vib_m1.median()
med_vib_m2 = vib_m2.median()
med_vib_geral = df['vibracao'].median()
resp_a = round(max(med_vib_m1, med_vib_m2), 2) if p_val_a < alpha else round(med_vib_geral, 2)

# ITEM B - Temperatura: teste t pareado
temp_antes = df['temperatura_antes']
temp_depois = df['temperatura_depois']
t_stat, p_val_b = stats.ttest_rel(temp_antes, temp_depois, alternative='two-sided')
media_antes = temp_antes.mean()
media_depois = temp_depois.mean()
media_geral_temp = pd.concat([temp_antes, temp_depois]).mean()
resp_b = round(min(media_antes, media_depois), 2) if p_val_b < alpha else round(media_geral_temp, 2)

# ITEM C - Temperatura: Wilcoxon pareado
w_stat, p_val_c = stats.wilcoxon(temp_antes, temp_depois, alternative='two-sided')
med_antes = temp_antes.median()
med_depois = temp_depois.median()
med_geral_temp = pd.concat([temp_antes, temp_depois]).median()
resp_c = round(min(med_antes, med_depois), 2) if p_val_c < alpha else round(med_geral_temp, 2)

# ITEM D - Defeitos: teste t independente
def_m1 = m1['defeitos']
def_m2 = m2['defeitos']
t_stat_d, p_val_d = stats.ttest_ind(def_m1, def_m2, alternative='two-sided')
media_def_m1 = def_m1.mean()
media_def_m2 = def_m2.mean()
media_def_geral = df['defeitos'].mean()
resp_d = round(max(media_def_m1, media_def_m2), 2) if p_val_d < alpha else round(media_def_geral, 2)

# DETALHES
print("=" * 60)
print("ITEM A - Vibração: Mann-Whitney U")
print("=" * 60)
print(f"Mediana vibração M1: {med_vib_m1:.4f}")
print(f"Mediana vibração M2: {med_vib_m2:.4f}")
print(f"Mediana geral:       {med_vib_geral:.4f}")
print(f"U={u_stat:.4f}, p={p_val_a:.4f} → {'Sig.' if p_val_a < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM B - Temperatura: Teste t pareado")
print("=" * 60)
print(f"Média temp antes:  {media_antes:.4f}")
print(f"Média temp depois: {media_depois:.4f}")
print(f"Média geral:       {media_geral_temp:.4f}")
print(f"t={t_stat:.4f}, p={p_val_b:.4f} → {'Sig.' if p_val_b < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM C - Temperatura: Wilcoxon pareado")
print("=" * 60)
print(f"Mediana temp antes:  {med_antes:.4f}")
print(f"Mediana temp depois: {med_depois:.4f}")
print(f"Mediana geral:       {med_geral_temp:.4f}")
print(f"W={w_stat:.4f}, p={p_val_c:.4f} → {'Sig.' if p_val_c < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM D - Defeitos: Teste t independente")
print("=" * 60)
print(f"Média defeitos M1: {media_def_m1:.4f}")
print(f"Média defeitos M2: {media_def_m2:.4f}")
print(f"Média geral:       {media_def_geral:.4f}")
print(f"t={t_stat_d:.4f}, p={p_val_d:.4f} → {'Sig.' if p_val_d < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("RESUMO FINAL")
print("=" * 60)
soma = resp_a + resp_b + resp_c + resp_d
print(f"a) {resp_a}")
print(f"b) {resp_b}")
print(f"c) {resp_c}")
print(f"d) {resp_d}")
print(f"\nSOMA TOTAL: {soma:.2f}")
```

```python
# Instancia 21
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv(r'C:\FIAP\Hipoteses_Nemec_09_03_26_Futebol\Dados\instancia_21_ecommerce\dataset.csv')
df.columns = df.columns.str.strip().str.lstrip('\ufeff')

alpha = 0.05

camp_a = df[df['campanha'] == 'A']
camp_b = df[df['campanha'] == 'B']
mobile  = df[df['dispositivo'] == 'mobile']
desktop = df[df['dispositivo'] == 'desktop']

# ============================================================
# ITEM A - Ticket médio: teste t independente (A vs B)
# ============================================================
t_stat_a, p_val_a = stats.ttest_ind(camp_a['ticket_medio'], camp_b['ticket_medio'], alternative='two-sided')
media_ticket_a    = camp_a['ticket_medio'].mean()
media_ticket_b    = camp_b['ticket_medio'].mean()
media_ticket_geral = df['ticket_medio'].mean()
resp_a = round(max(media_ticket_a, media_ticket_b), 2) if p_val_a < alpha else round(media_ticket_geral, 2)

# ============================================================
# ITEM B - Tempo antes: Welch t-test (mobile vs desktop)
# ============================================================
t_stat_b, p_val_b = stats.ttest_ind(mobile['tempo_antes_min'], desktop['tempo_antes_min'],
                                     alternative='two-sided', equal_var=False)
media_tempo_mobile  = mobile['tempo_antes_min'].mean()
media_tempo_desktop = desktop['tempo_antes_min'].mean()
media_tempo_geral   = df['tempo_antes_min'].mean()
resp_b = round(max(media_tempo_mobile, media_tempo_desktop), 2) if p_val_b < alpha else round(media_tempo_geral, 2)

# ============================================================
# ITEM C - Conversão: Wilcoxon pareado (antes vs depois)
# ============================================================
w_stat, p_val_c = stats.wilcoxon(df['conversao_antes_pct'], df['conversao_depois_pct'], alternative='two-sided')
med_conv_antes  = df['conversao_antes_pct'].median()
med_conv_depois = df['conversao_depois_pct'].median()
med_conv_geral  = pd.concat([df['conversao_antes_pct'], df['conversao_depois_pct']]).median()
resp_c = round(max(med_conv_antes, med_conv_depois), 2) if p_val_c < alpha else round(med_conv_geral, 2)

# ============================================================
# ITEM D - Satisfação: Mann-Whitney U (A vs B)
# ============================================================
u_stat, p_val_d = stats.mannwhitneyu(camp_a['satisfacao'], camp_b['satisfacao'], alternative='two-sided')
med_sat_a    = camp_a['satisfacao'].median()
med_sat_b    = camp_b['satisfacao'].median()
med_sat_geral = df['satisfacao'].median()
resp_d = round(max(med_sat_a, med_sat_b), 2) if p_val_d < alpha else round(med_sat_geral, 2)

# ============================================================
# DETALHES
# ============================================================
print("=" * 60)
print("ITEM A - Ticket médio: Teste t independente (A vs B)")
print("=" * 60)
print(f"Média ticket A:    {media_ticket_a:.4f}")
print(f"Média ticket B:    {media_ticket_b:.4f}")
print(f"Média geral:       {media_ticket_geral:.4f}")
print(f"t={t_stat_a:.4f}, p={p_val_a:.4f} → {'Sig.' if p_val_a < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM B - Tempo antes: Welch t-test (mobile vs desktop)")
print("=" * 60)
print(f"Média tempo mobile:  {media_tempo_mobile:.4f}")
print(f"Média tempo desktop: {media_tempo_desktop:.4f}")
print(f"Média geral:         {media_tempo_geral:.4f}")
print(f"t={t_stat_b:.4f}, p={p_val_b:.4f} → {'Sig.' if p_val_b < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM C - Conversão: Wilcoxon pareado (antes vs depois)")
print("=" * 60)
print(f"Mediana conversão antes:  {med_conv_antes:.4f}")
print(f"Mediana conversão depois: {med_conv_depois:.4f}")
print(f"Mediana geral:            {med_conv_geral:.4f}")
print(f"W={w_stat:.4f}, p={p_val_c:.4f} → {'Sig.' if p_val_c < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM D - Satisfação: Mann-Whitney U (A vs B)")
print("=" * 60)
print(f"Mediana satisfação A: {med_sat_a:.4f}")
print(f"Mediana satisfação B: {med_sat_b:.4f}")
print(f"Mediana geral:        {med_sat_geral:.4f}")
print(f"U={u_stat:.4f}, p={p_val_d:.4f} → {'Sig.' if p_val_d < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("RESUMO FINAL")
print("=" * 60)
soma = resp_a + resp_b + resp_c + resp_d
print(f"a) {resp_a}")
print(f"b) {resp_b}")
print(f"c) {resp_c}")
print(f"d) {resp_d}")
print(f"\nSOMA TOTAL: {soma:.2f}")
```

```python
# Instancia 22
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv(r'C:\FIAP\Hipoteses_Nemec_09_03_26_Futebol\Dados\instancia_22_manufatura\dataset.csv')
df.columns = df.columns.str.strip().str.lstrip('\ufeff')

alpha = 0.05

m1   = df[df['maquina'] == 'M1']
m2   = df[df['maquina'] == 'M2']
dia  = df[df['turno'] == 'dia']
noite = df[df['turno'] == 'noite']

# ============================================================
# ITEM A - Temperatura: Wilcoxon pareado (antes vs depois)
# ============================================================
w_stat, p_val_a = stats.wilcoxon(df['temperatura_antes'], df['temperatura_depois'], alternative='two-sided')
med_antes  = df['temperatura_antes'].median()
med_depois = df['temperatura_depois'].median()
med_geral  = pd.concat([df['temperatura_antes'], df['temperatura_depois']]).median()
resp_a = round(min(med_antes, med_depois), 2) if p_val_a < alpha else round(med_geral, 2)

# ============================================================
# ITEM B - Temperatura: Teste t pareado (antes vs depois)
# ============================================================
t_stat_b, p_val_b = stats.ttest_rel(df['temperatura_antes'], df['temperatura_depois'], alternative='two-sided')
media_antes  = df['temperatura_antes'].mean()
media_depois = df['temperatura_depois'].mean()
media_geral_temp = pd.concat([df['temperatura_antes'], df['temperatura_depois']]).mean()
resp_b = round(min(media_antes, media_depois), 2) if p_val_b < alpha else round(media_geral_temp, 2)

# ============================================================
# ITEM C - Defeitos: Teste t independente (M1 vs M2)
# ============================================================
t_stat_c, p_val_c = stats.ttest_ind(m1['defeitos'], m2['defeitos'], alternative='two-sided')
media_def_m1   = m1['defeitos'].mean()
media_def_m2   = m2['defeitos'].mean()
media_def_geral = df['defeitos'].mean()
resp_c = round(max(media_def_m1, media_def_m2), 2) if p_val_c < alpha else round(media_def_geral, 2)

# ============================================================
# ITEM D - Tempo de ciclo: Welch t-test (dia vs noite)
# ============================================================
t_stat_d, p_val_d = stats.ttest_ind(dia['tempo_ciclo_s'], noite['tempo_ciclo_s'],
                                     alternative='two-sided', equal_var=False)
media_tempo_dia   = dia['tempo_ciclo_s'].mean()
media_tempo_noite = noite['tempo_ciclo_s'].mean()
media_tempo_geral = df['tempo_ciclo_s'].mean()
resp_d = round(max(media_tempo_dia, media_tempo_noite), 2) if p_val_d < alpha else round(media_tempo_geral, 2)

# ============================================================
# DETALHES
# ============================================================
print("=" * 60)
print("ITEM A - Temperatura: Wilcoxon pareado")
print("=" * 60)
print(f"Mediana temp antes:  {med_antes:.4f}")
print(f"Mediana temp depois: {med_depois:.4f}")
print(f"Mediana geral:       {med_geral:.4f}")
print(f"W={w_stat:.4f}, p={p_val_a:.4f} → {'Sig.' if p_val_a < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM B - Temperatura: Teste t pareado")
print("=" * 60)
print(f"Média temp antes:  {media_antes:.4f}")
print(f"Média temp depois: {media_depois:.4f}")
print(f"Média geral:       {media_geral_temp:.4f}")
print(f"t={t_stat_b:.4f}, p={p_val_b:.4f} → {'Sig.' if p_val_b < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM C - Defeitos: Teste t independente (M1 vs M2)")
print("=" * 60)
print(f"Média defeitos M1: {media_def_m1:.4f}")
print(f"Média defeitos M2: {media_def_m2:.4f}")
print(f"Média geral:       {media_def_geral:.4f}")
print(f"t={t_stat_c:.4f}, p={p_val_c:.4f} → {'Sig.' if p_val_c < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("ITEM D - Tempo ciclo: Welch t-test (dia vs noite)")
print("=" * 60)
print(f"Média tempo dia:   {media_tempo_dia:.4f}")
print(f"Média tempo noite: {media_tempo_noite:.4f}")
print(f"Média geral:       {media_tempo_geral:.4f}")
print(f"t={t_stat_d:.4f}, p={p_val_d:.4f} → {'Sig.' if p_val_d < alpha else 'Não sig.'}")

print("\n" + "=" * 60)
print("RESUMO FINAL")
print("=" * 60)
soma = resp_a + resp_b + resp_c + resp_d
print(f"a) {resp_a}")
print(f"b) {resp_b}")
print(f"c) {resp_c}")
print(f"d) {resp_d}")
print(f"\nSOMA TOTAL: {soma:.2f}")
```
